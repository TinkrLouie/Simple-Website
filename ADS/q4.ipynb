```python
def balanced(n):

    def permutations(iterable, r=None):
        # structure of this function is closely modeled after permutations of itertools
        pool = tuple(iterable)
        d = len(pool)
        r = d if r is None else r
        if r > d:
            return
        indices = list(range(d))
        cycles = list(range(d, d - r, -1))
        yield tuple(pool[i] for i in indices[:r])
        while d:
            for i in reversed(range(r)):
                cycles[i] -= 1
                if cycles[i] == 0:
                    indices[i:] = indices[i + 1:] + indices[i:i + 1]
                    cycles[i] = d - i
                else:
                    j = cycles[i]
                    indices[i], indices[-j] = indices[-j], indices[i]
                    yield tuple(pool[i] for i in indices[:r])
                    break
            else:
                return

    # function below recursively calls permutations
    def unique_perms(perm):
        return {"".join(p) for p in permutations(perm)}

    counter = 0

    if n <= 0:
        print('Invalid input')
        return
    if 1 <= n <= 2:
        return 0
    a = []
    for x in range(1, n + 1):
        x = str(x)
        a.append(x)
        a.append(x)

    for c in sorted(unique_perms(a)):
        count = 0
        index = 0
        for k in c:
            k_i = index
            nd_k_i = k_i + int(k) + 1
            index += 1
            if nd_k_i > (2 * n - 1):
                continue
            if c[k_i] == c[nd_k_i]:
                count += 1
            else:
                break

        if count == n:
            counter += 1

    print(counter)

    return counter
```
